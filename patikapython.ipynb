Liste Düzleştirme

```
a =  [[10,11,12],[13,14],[15,16,17,18]]
def duzlestırme(a):
    for l in a:
        print()
    yenı_a = []
    for l in a:
        for e in l:
            yenı_a.append(e)
        print(yenı_a)
duzlestırme(a)
```
